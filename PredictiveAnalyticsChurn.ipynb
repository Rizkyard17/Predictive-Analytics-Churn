# EDA

```python
excuse = ['customerID']

for i in df.columns:
  if i not in excuse:
    print('=============================')
    print(df[i].value_counts())
```

```python
excuse = ['customerID', 'MonthlyCharges', 'TotalCharges']

for i in df.columns:
  if i not in excuse:

    plt.figure(figsize = (10,5))
    sns.countplot(x=i, data=df, palette='Set3', order=df[i].value_counts().index)
    plt.title('Jumlah {}'.format(i))
    plt.xlabel(i)
    plt.ylabel('Jumlah')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
